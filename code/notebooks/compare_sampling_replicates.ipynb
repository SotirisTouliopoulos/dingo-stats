# Melt the correlation matrices of the replicates

```python


corr_matrix_001 = correlated_reactions(
                            absolute_steady_states_001,
                            pearson_cutoff = 0.8,
                            indicator_cutoff = 0,
                            lower_triangle = False,
                            )

corr_matrix_001[np.isnan(corr_matrix_001)] = 0

corr_matrix_melted_001 = corr_matrix_001[np.triu_indices_from(corr_matrix_001, k=1)]



corr_matrix_002 = correlated_reactions(
                            absolute_steady_states_002,
                            pearson_cutoff = 0.8,
                            indicator_cutoff = 0,
                            lower_triangle = False,
                            )

corr_matrix_002[np.isnan(corr_matrix_002)] = 0

corr_matrix_melted_002 = corr_matrix_002[np.triu_indices_from(corr_matrix_002, k=1)]



corr_matrix_003 = correlated_reactions(
                            absolute_steady_states_003,
                            pearson_cutoff = 0.8,
                            indicator_cutoff = 0,
                            lower_triangle = False,
                            )

corr_matrix_003[np.isnan(corr_matrix_003)] = 0

corr_matrix_melted_003 = corr_matrix_003[np.triu_indices_from(corr_matrix_003, k=1)]


```

# Calculate correlation between melted correlation matrices to see if we have reproducibility across replicates

```python

from scipy.stats import pearsonr
from scipy.stats import spearmanr

# For optimality = 0.01%
# When giving filtered correlation matrix with 0.8 cutoff we get ==> 0.92, 0.94, 0.87, 0.89, 0.92, 0.83

# For optimality = 1%
# When giving filtered correlation matrix with 0.8 cutoff we get ==> 0.99, 0.89, 0.90, 0.99, 0.87, 0.87

# For optimality = 20%
# When giving filtered correlation matrix with 0.8 cutoff we get ==> 0.99

# For optimality = 40%
# When giving filtered correlation matrix with 0.8 cutoff we get ==> 0.99, 0.98, 0.99, 0.98, 0.98, 0.99

# For optimality = 60%
# When giving filtered correlation matrix with 0.8 cutoff we get ==> 0.99, 0.99, 0.99, 0.98, 0.99, 0.98

# For optimality = 80%
# When giving filtered correlation matrix with 0.8 cutoff we get ==> 0.99

# For optimality = 99%
# When giving filtered correlation matrix with 0.8 cutoff we get ==>  0.99

# For optimality = 100% with subsamples
# When giving filtered correlation matrix with 0.8 cutoff we get ==> 0.32, 0.52, 0.27


corr, p_value = pearsonr(corr_matrix_melted_001, corr_matrix_melted_002)
print(corr)
corr, p_value = pearsonr(corr_matrix_melted_001, corr_matrix_melted_003)
print(corr)
corr, p_value = pearsonr(corr_matrix_melted_002, corr_matrix_melted_003)
print(corr)


corr, p = spearmanr(corr_matrix_melted_001, corr_matrix_melted_002)
print(corr)
corr, p = spearmanr(corr_matrix_melted_001, corr_matrix_melted_003)
print(corr)
corr, p = spearmanr(corr_matrix_melted_002, corr_matrix_melted_003)
print(corr)
```
