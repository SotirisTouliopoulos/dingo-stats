```python

conditions = [
    absolute_steady_states_0_001,
    absolute_steady_states_0_002,
    absolute_steady_states_0_003,
    
    absolute_steady_states_1_001,
    absolute_steady_states_1_002,
    absolute_steady_states_1_003,
    
    absolute_steady_states_99_001,
    absolute_steady_states_99_002,
    absolute_steady_states_99_003
]


# Add a fold change ?


selected_comparisons = [(0, 1), (0, 2), (1, 2),
                        (3, 4), (3, 5), (4, 5),
                        (6, 7), (6, 8), (7, 8)
                        ]

# Store p-values for each row for KS test
p_values_ks = {row: [] for row in range(len(reactions)) }
# Store ks-values for each row for KS test
values_ks = {row: [] for row in range(len(reactions)) }

for row in range(len(reactions)):
    for i, j in selected_comparisons:
        
        ks, p_ks = stats.ks_2samp(conditions[i][row], conditions[j][row], alternative='two-sided')

        p_values_ks[row].append(p_ks)
        values_ks[row].append(ks)
        
        print(row, i, j, ks, p_ks)

"""
# How p_values list looks like

p_values = {
    0: [0.0, 0.0], # Reaction 1: Condition 0 vs 1, Condition 0 vs 2
    1: [0.0, 0.0],
    2: [0.0, 0.0],
    3: [0.0, 0.0]
}
"""

```

# Correct p values for multiple testing

```python

print(p_values_ks)
p_values_ks_copy = list(p_values_ks.values())
flat_p_values_ks = np.array(p_values_ks_copy).flatten()
print(flat_p_values_ks)

# Apply FDR correction
rejected, corrected_p_values_ks, _, _ = multipletests(flat_p_values_ks, method='fdr_bh')

# Reshape corrected p-values back to the original matrix shape
corrected_p_values_ks = corrected_p_values_ks.reshape(np.array(p_values_ks_copy).shape)
print(corrected_p_values_ks)



raw_p = [p for row in p_values_ks.values() for p in row]
corrected_p = corrected_p_values_ks.flatten()

plt.scatter(raw_p, corrected_p)
plt.xlabel("Raw p-values")
plt.ylabel("Corrected p-values")
plt.title("Raw vs. Corrected p-values")
#plt.xscale("log")
#plt.yscale("log")
plt.show()
```

# Melt the correlation matrices of the replicates

```python


corr_matrix_001 = correlated_reactions(
                            absolute_steady_states_1_001,
                            pearson_cutoff = 0.8,
                            indicator_cutoff = 0,
                            lower_triangle = False,
                            )

corr_matrix_001[np.isnan(corr_matrix_001)] = 0

corr_matrix_melted_001 = corr_matrix_001[np.triu_indices_from(corr_matrix_001, k=1)]



corr_matrix_002 = correlated_reactions(
                            absolute_steady_states_1_002,
                            pearson_cutoff = 0.8,
                            indicator_cutoff = 0,
                            lower_triangle = False,
                            )

corr_matrix_002[np.isnan(corr_matrix_002)] = 0

corr_matrix_melted_002 = corr_matrix_002[np.triu_indices_from(corr_matrix_002, k=1)]



corr_matrix_003 = correlated_reactions(
                            absolute_steady_states_1_003,
                            pearson_cutoff = 0.8,
                            indicator_cutoff = 0,
                            lower_triangle = False,
                            )

corr_matrix_003[np.isnan(corr_matrix_003)] = 0

corr_matrix_melted_003 = corr_matrix_003[np.triu_indices_from(corr_matrix_003, k=1)]


```
